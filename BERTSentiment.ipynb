```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
# import requests
# from bs4 import BeautifulSoup
import re
```

2. insantiate model
    * the model that used : https://huggingface.co/nlptown/bert-base-multilingual-uncased-sentiment

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

3. encode and calculate sentiment <br>
input review here!

```python
review2 = "Pretty disappointing. Was told this was one of the highest rated bars in Asia. Waiter didn't even know what a Jameson and coke was along with Hennessy pineapple. Not one star due to the fact that I enjoyed the live music and the decor."
review3 = "One of my favorite bars in Taipei. Solid drinks menu with some very unique selections for a fairly compact selection. They're located in the same building as MQ so you can also bounce between the two. It's a lot different vibe though. You are guided up towards a set of stairs and through the back of a bookshelf into the classy speakeasy. Usually some good jazz in there and sometimes a live ensemble. Weekends they do require a two drink minimum so be ready for about 900-1000NT charge."

tokens1 = tokenizer.encode('I hated this, absolutely the worst', return_tensors='pt')
tokens2 = tokenizer.encode(review2, return_tensors='pt')
tokens3 = tokenizer.encode(review3, return_tensors='pt')

tokens_list = [tokens1, tokens2, tokens3]

for tokens in tokens_list:
    print(tokens)
```

```python
for tokens in tokens_list:
    print(tokenizer.decode(tokens[0]))
```

The output from the model is a one-hot encoded list of scores.
The position with the highest score represents the sentiment roting e.g. [0.9,0.2,0.1,-0.2,-0.5] is a roting of 1.

```python
result_list = []
i=0
comment = "review "

for tokens in tokens_list:
    i+=1
    result = model(tokens)
    print("\n"+comment+str(i)+":")
    print(result.logits)
    rating = int(torch.argmax(result.logits))+1
    print("rating score:"+ str(rating))
```
